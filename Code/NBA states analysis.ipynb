```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
all_players = pd.read_csv(r"C:\Users\moham\Downloads\Projects\NBA States Analysis\Dataset\Player Per Game.csv")
```

```python
performance_columns = [
    'pts_per_game', 'ast_per_game', 'trb_per_game',
    'stl_per_game', 'blk_per_game', 'fg_percent'
]

performance_data = all_players[performance_columns]

# Calculate mean performance for each column
mean_performance = performance_data.mean()
```

```python
mean_performance
```

```python
plt.figure(figsize=(10, 6))
mean_performance.plot(kind='bar', color='skyblue')
plt.title('Average Player Performance')
plt.xlabel('Performance Metrics')
plt.ylabel('Average Value')
plt.show()
```

```python
# Player Age Trends
plt.figure(figsize=(10, 6))
sns.boxplot(x='age', y='pts_per_game', data=all_players, palette='viridis')
plt.title('Player Age vs. Points Per Game')
plt.xlabel('Age')
plt.ylabel('Points Per Game')
plt.show()
```

```python
selected_player = 'Stephen Curry'
player_data = all_players[all_players['player'] == selected_player]

# Player Performance Trends
plt.figure(figsize=(12, 8))
plt.plot(player_data['season'], player_data['pts_per_game'], marker='o', label='Points Per Game', color='blue')
plt.plot(player_data['season'], player_data['ast_per_game'], marker='o', label='Assists Per Game', color='green')
plt.plot(player_data['season'], player_data['trb_per_game'], marker='o', label='Rebounds Per Game', color='orange')

plt.title(f'{selected_player} Performance Trends Over Seasons')
plt.xlabel('Season')
plt.ylabel('Performance Metrics')
plt.legend()
plt.show()

# Player Age Trends
plt.figure(figsize=(8, 6))
sns.lineplot(x='age', y='pts_per_game', data=player_data, marker='o', color='purple')
plt.title(f'{selected_player} Age vs. Points Per Game')
plt.xlabel('Age')
plt.ylabel('Points Per Game')
plt.show()
```

```python
player_shooting = pd.read_csv(r"C:\Users\moham\Downloads\Projects\NBA States Analysis\Dataset\Player Shooting.csv")
```

```python
# Select a specific player
selected_player = 'Al Horford'
player_shooting_data = player_shooting[player_shooting["player"] == selected_player]
```

```python
# Shooting Efficiency Trends
plt.figure(figsize=(12, 8))
plt.plot(player_shooting_data['season'], player_shooting_data['fg_percent'], marker='o', label='Field Goal Percentage', color='blue')
plt.plot(player_shooting_data['season'], player_shooting_data['fg_percent_from_x2p_range'], marker='o', label='2-Point Field Goal Percentage', color='green')
plt.plot(player_shooting_data['season'], player_shooting_data['fg_percent_from_x3p_range'], marker='o', label='3-Point Field Goal Percentage', color='orange')

plt.title(f'{selected_player} Shooting Efficiency Trends Over Seasons')
plt.xlabel('Season')
plt.ylabel('Field Goal Percentage')
plt.legend()
plt.show()
```

```python
# Shot Distribution
plt.figure(figsize=(12, 8))
sns.scatterplot(x='avg_dist_fga', y='percent_fga_from_x2p_range', data=player_shooting_data, hue='season', palette='viridis')
plt.title(f'{selected_player} Shot Distribution and Average Distance Over Seasons')
plt.xlabel('Average Shot Distance')
plt.ylabel('Percentage of Shots from 2-Point Range')
plt.legend()
plt.show()
```

```python
player_100poss = pd.read_csv(r"C:\Users\moham\Downloads\Projects\NBA States Analysis\Dataset\Per 100 Poss.csv")
```

```python
# Exclude non-numeric columns
numeric_cols = player_100poss.select_dtypes(include=['float64', 'int64']).columns
numeric_data = player_100poss[numeric_cols]
```

```python
print(numeric_data.head())
```

```python
# Basic statistics summary
summary_stats = numeric_data.describe()
print('\nSummary Statistics:')
print(summary_stats)
```

```python
# Visualize the distribution of key performance metrics
plt.figure(figsize=(12, 8))
sns.boxplot(x='pos', y='pts_per_100_poss', data=player_100poss, palette='viridis')
plt.title('Points per 100 Possessions by Position')
plt.show()
```

```python
# Explore correlations between variables
correlation_matrix = numeric_data.corr()
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Compare offensive and defensive ratings
plt.figure(figsize=(10, 6))
sns.scatterplot(x='o_rtg', y='d_rtg', data=player_100poss, hue='pos', palette='viridis', alpha=0.7)
plt.title('Offensive vs Defensive Rating')
plt.xlabel('Offensive Rating')
plt.ylabel('Defensive Rating')
plt.show()
```
